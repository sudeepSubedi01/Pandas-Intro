# Problem Statement
<b>Create a dataset with two columns: season, orangecap holder in each season</b>

```python
import pandas as pd
```

```python
match = pd.read_csv('matches.csv')
delivery = pd.read_csv('deliveries_new.csv')
```

```python
match
```

```python
delivery
```

```python
df = delivery.merge(match,left_on='match_id', right_on='id')
df.columns
```

```python
df.groupby(['batter'])['batsman_runs'].sum()
```

```python
new_df = df.groupby(['season','batter'])['batsman_runs'].sum().sort_values(ascending=False).reset_index()
new_df
```

```python
df2 = new_df.drop_duplicates('season',keep='first').sort_values('season')
df2
```

```python
df2.drop('batsman_runs',axis=1, inplace=True)
```

```python
df2.rename(columns={'batter':'orange_cap_holder'})
```
