```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame()
df['X1'] = [1,2,3,4,5,6,6,7,9,9]
df['X2'] = [5,3,6,8,1,9,5,8,9,2]
df['label'] = [1,1,0,1,0,1,0,1,0,0]
df
```

```python
from sklearn.tree import DecisionTreeClassifier
dt = DecisionTreeClassifier(max_depth=1)
# dt.fit(df[].reshape(10,1),df['label'].reshape(10,1))
dt.fit(df.iloc[:,0:2],df.iloc[:,2])
y_pred = dt.predict(df.iloc[:,0:2])

df['weights'] = 1 / df.shape[0]
df['y_pred'] = y_pred
df
```

```python
def update_row_weights(rowaaa, alpha=0.42364893019360184):        # each row is passed to this function
  if rowaaa['label'] == rowaaa['y_pred']:
    return rowaaa['weights'] * np.exp(-alpha)
  else:
    return rowaaa['weights'] * np.exp(alpha)
```

```python
df.apply(update_row_weights,axis=1) 
```

```python
df['updated_weights'] = df.apply(update_row_weights,axis=1) 
df
```
