# Finding out against which team V Kohli scored most runs

```python
new_df = df[df['batsman']=='V Kohli']
new_df.head()
```

```python
against = new_df.groupby('bowling_team')
against['batsman_runs'].sum().sort_values(ascending=False)
```

# Finding out against which team A PLAYER scored most runs

```python
def runs_scored(player):
  new_df = df[df['batsman']==player]
  new_df.head()
  against = new_df.groupby('bowling_team')
  print(against['batsman_runs'].sum().sort_values(ascending=False).head())
```

```python
runs_scored('G Gambhir')
```
