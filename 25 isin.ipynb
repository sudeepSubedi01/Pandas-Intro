# Problem Statement<br>
<b>Find the most destructive death over batsman in the history of IPL</b>
<br>
<br>
<br>
<b>Solution:</b><br>
- Choosing only those batsman who has played more than 200 balls in death over in all the matches <br>
- Calculate strike rate and then decide <br>
srtike rate = number of runs / number of balls * 100

```python
import pandas as pd
df = pd.read_csv('deliveries.csv')
df
```

```python
death_over_df = df[df['over'] >= 16]
death_over_df
```

```python
# since each row corresponds to one ball played by a player, this returns total balls played by the player in 16-20 overs
eligible_batsman = death_over_df.groupby('batsman')['batsman_runs'].count()
eligible_batsman = eligible_batsman[eligible_batsman.values > 200]
eligible_batsman
```

```python
type(eligible_batsman.index)
```

```python
batsman_list = list(eligible_batsman.index)
batsman_list
```

```python
final = death_over_df[death_over_df['batsman'].isin(batsman_list)]
final
```

```python
balls = pd.DataFrame(final.groupby('batsman').count()['ball'].reset_index())
balls
```

```python
runs = pd.DataFrame(death_over_df.groupby('batsman')['batsman_runs'].sum().reset_index())
```

```python
runs_balls = runs.merge(balls, on=['batsman'])
runs_balls
```

```python
runs_balls['strike_rate'] = runs_balls['batsman_runs'] / runs_balls['ball'] * 100
runs_balls.sort_values('strike_rate',ascending=False).head()
```
